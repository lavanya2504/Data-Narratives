```python
import pandas as pd
import matplotlib.pyplot as plt
df1=pd.read_csv("/content/book_tags.csv")
df2=pd.read_csv("/content/books.csv")
df3=pd.read_csv("/content/ratings.csv")
df4=pd.read_csv("/content/tags.csv")
df5=pd.read_csv("/content/to_read.csv")
```

The probability that if a book is in the top 10 most recommended books list,then it is also in the top 10 books having highest average rating. 


```python

# count the number of occurrences of each book_id in df1
book_id_counts = df5['book_id'].value_counts(ascending=False)

# select the top 10 most occurring book_ids
top_10_book_ids = book_id_counts.index[:10]

new_df=df2[df2["book_id"].isin(top_10_book_ids)]
print(new_df[["original_title","average_rating"]])

ax=new_df.plot(x="original_title",y="average_rating",kind="line",xticks=range(len(new_df['original_title'])),title="Most preferred books v/s average rating")
ax.set_xticklabels(new_df['original_title'], rotation=60)

print()
new_df2=(df2.sort_values(by="average_rating",ascending=False)).head(10)
print(new_df2[["original_title","average_rating"]])

```

to find and plot top 15 book pulishing years and the number of books published in those years

```python
df_=(df2["original_publication_year"].value_counts(ascending=False)).head(15)
new_df=pd.DataFrame({"year":df_.index,"count":df_.values})
new_df1=new_df.sort_values(by="year")
ax=new_df1.plot(kind="bar",x="year",y="count")
ax.set_xlabel("publication year")
ax.set_ylabel("number of books published")
```

to find top 10 authors by number of books written by them having average rating greater than 4

```python
df2_=df2[df2["average_rating"]>4]
df__=(df2["authors"].value_counts(ascending=False)).head(10)
new_df=pd.DataFrame({"Author":df__.index,"number of books":df__.values})
print(new_df)
ax=new_df.plot(x="Author",y="number of books",kind="bar")
ax.set_ylabel("number of books")
```
